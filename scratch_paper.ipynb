```python
import numpy as np
import matplotlib.pyplot as plt

# Generate angles from 0 to 2*pi (0 to 360 degrees) with small increments
theta = np.linspace(0, 2*np.pi, 1000)

# Calculate x and y coordinates of points on the unit circle
x = np.cos(theta)
y = np.sin(theta)

# Calculate points for sine, cosine, and tangent lines
angles = [np.pi/4, np.pi/3, np.pi/6]  # Example angles in radians (45, 60, and 30 degrees)
sin_points = np.sin(angles)
cos_points = np.cos(angles)
tan_points = np.tan(angles)

# Plot the unit circle
plt.figure(figsize=(6, 6))
plt.plot(x, y, label='Unit Circle')

# Plot points for sine, cosine, and tangent lines
plt.scatter(cos_points, sin_points, color='red', label='Cosine & Sine Points')
for i in range(len(angles)):
    plt.plot([0, cos_points[i]], [0, sin_points[i]], color='red', linestyle='--')
    plt.plot([cos_points[i], cos_points[i]], [0, sin_points[i]], color='red', linestyle='--')
    plt.annotate(f'({cos_points[i]:.2f}, {sin_points[i]:.2f})', (cos_points[i], sin_points[i]),
                 textcoords="offset points", xytext=(10,10), ha='center')

plt.plot([0, 1], [0, 0], color='blue', label='cos axis')
plt.plot([0, 0], [0, 1], color='green', label='sin axis')

# Set plot limits and labels
plt.xlim([-1.2, 1.2])
plt.ylim([-1.2, 1.2])
plt.xlabel('cos')
plt.ylabel('sin')
plt.title('Unit Circle with Sine, Cosine, and Tangent Lines')

# Move the legend to the upper-right corner
plt.legend(loc='upper left')

plt.grid(True)
plt.show()
```

```python
# Define the function
def f(x):
    return (x**2 - 1) / (x - 1)

# Define the x values for the plot
x = np.linspace(0.5, 1.5, 400)

# Calculate the function values
y = f(x)

# Plot the function
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$\frac{x^2 - 1}{x - 1}$', color='b')

# Draw the limit line (horizontal line at y=2)
plt.axhline(y=2, color='r', linestyle='--', label='Limit as x approaches 1')

# Mark the point where x approaches 1
plt.plot(1, 2, 'ro')  # Red dot at (1, 2)

# Set plot labels and title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Visualization of the Limit')
plt.legend()
plt.grid(True)
plt.show()
```
