# Introduction to Limits in Calculus

The concept of **limits** plays a fundamental role in Calculus. Limits allow us to understand the behavior of functions as they approach specific points or values. 

Consider a function $f(x)$. The limit of $f(x)$ as $x$ approaches a certain value, say $a$, represents the value that $f(x)$ approaches as $x$ gets arbitrarily close to $a$.

Mathematically, we denote the limit as:

$$\lim_{{x \to a}} f(x)$$

This signifies the value that $f(x)$ approaches as $x$ approaches $a$.

Limits are essential for analyzing continuity, derivatives, and integrals in calculus, forming the basis for many advanced concepts in mathematics.

### Different kinds of limits

Limits can be straightforward, where $\lim_{{x \to a}} f(x) = f(a)$. See the below example where we have the function $f(x) = 2x - 5$. The $\lim_{{x \to 5}} f(x) = f(5) = 5$. As $x$ gets closer and closer to $5$ from the right and left, we see that $y$ approaches $5$, which is equivalent to $f(5)$. 

```python
# Define the function f(x) = 2x - 5
def f(x):
    return 2 * x - 5

# Generate x values
x = np.linspace(-10, 10, 400)  # 400 points between -10 and 10
# Generate y values using the function
y = f(x)
# Plot the function
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = 2x - 5', color='blue')
plt.scatter(5, 5, s=49, c='black')
plt.arrow(3, 2, 1, 2, head_width=0.3, head_length=1, 
          fc='red', ec='red')
plt.arrow(7, 10, -1, -2, head_width=0.3, head_length=1, 
          fc='red', ec='red')
# Add labels and title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(r'Depiction of $\lim_{{x \to 5}} \: f(x)$')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')  # x-axis
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')  # y-axis
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

However, limits are not always so straightforward. There are instances where $\lim_{{x \to a}} f(x) \neq f(a)$. See the below example where we have the function $f(x) = \frac{x^2 - 1}{x - 1}$ and the $\lim_{{x \to 1}} f(x)$. When solving for $f(1)$, we get the indeterminate form $\frac{0}{0}$ (good explanation of indeterminate forms [here](https://www.khanacademy.org/math/algebra/x2f8bb11595b61c86:foundation-algebra/x2f8bb11595b61c86:division-zero/v/undefined-and-indeterminate)). $1$ is the only value that will give us $\frac{0}{0}$, but any other value approaching $1$ from either side will give a real number. One potential method to find a limit of a rational function is to simplify:

$ \frac{x^2 - 1}{x - 1} = \frac{(x-1)(x+1)}{x-1}$
<br><br>
$\frac{(x-1)(x+1)}{x-1} = x+1$

When plugging $1$ in for $x+1$, we get $2$.

Thus, $f(1)$ is indeterminate, but the $\lim_{{x \to 1}} f(x) = 2$

```python
# Define the function with a removable discontinuity at x=1
def f(x):
    return (x**2 - 1) / (x - 1)

# Generate x values (excluding 1 to avoid division by zero)
x_left = np.linspace(-2, 0.95, 400)  # Values slightly less than 1
x_right = np.linspace(1.05, 2, 400)   # Values slightly greater than 1
x = np.concatenate((x_left, x_right))

# Generate y values using the function
y = f(x)

# Plot the function
plt.figure(figsize=(8, 6))
plt.plot(x_left, f(x_left), color='blue')
plt.plot(x_right, f(x_right), color='blue') 

# Add a blank circle at the gap (x=1)
plt.scatter(1, 2, color='black', facecolors='none', s=100, 
            edgecolors='black', label='Removable Discontinuity')

plt.arrow(0.6, 1.8, 0.2, 0.2, head_width=0.1, head_length=0.1, 
          fc='red', ec='red')
plt.arrow(1.4, 2.6, -0.2, -0.2, head_width=0.1, head_length=0.1, 
          fc='red', ec='red')

plt.text(-2, 4, r'$f(x) = \frac{x^2 - 1}{x - 1}$', fontsize=12, 
         bbox=dict(facecolor='white', alpha=0.3))


# Add labels and title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(r'Depiction of $\lim_{{x \to 1}} \: f(x)$ (removable discontinuity at x = 1)')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')  # x-axis
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')  # y-axis
plt.grid(True, linestyle='--', alpha=0.7)
plt.ylim(-5, 5)  # Set y-axis limits for better visualization
plt.show()
```

To further illustrate this distinction between $\lim_{{x \to a}} f(x)$ and $f(a)$, read the below code that uses $f(x)$ from the last example. As $x$ gets closer and closer to $1$ from the left and right, f gets closer and closer to $2$, but it will never arrive at $2$ until $x$ actually equals $1$. When dealing with limits, we care about what $f$ is approaching as $x$ approaches an arbitrary number $a$, not $f(a)$. Sometimes these things are equal and sometimes they aren't as we've observed above.

```python
for x in [0.9,0.99,0.999]:
    print(f'x = {x}, f(x) = {np.round(f(x),3)}')
print(50*'-')
for x in [1.1,1.01,1.001]:
    print(f'x = {x}, f(x) = {np.round(f(x),3)}')
```

### A Limit Must Be the Same from Both Sides

In the past two examples, we see the limit as $x$ approaches $a$ is the same coming from the left and right sides of the graph ($5$ in first example and $2$ in the second). If a function's limit from the left side isn't the same as the limit from the right side, then that limit doesn't exist.
