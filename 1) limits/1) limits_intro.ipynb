# Introduction to Limits in Calculus

The concept of **limits** plays a fundamental role in Calculus. Limits allow us to understand the behavior of functions as they approach specific points or values. 

Consider a function $f(x)$. The limit of $f(x)$ as $x$ approaches a certain value, say $a$, represents the value that $f(x)$ approaches as $x$ gets arbitrarily close to $a$.

Mathematically, we denote the limit as:

$$\lim_{{x \to a}} f(x)$$

This signifies the value that $f(x)$ approaches as $x$ approaches $a$.

Limits are essential for analyzing continuity, derivatives, and integrals in calculus, forming the basis for many advanced concepts in mathematics.

### Different kinds of limits

Limits can be straightforward, where $\lim_{{x \to a}} f(x) = f(a)$. See the below example where we have the function $f(x) = 2x - 5$. The $\lim_{{x \to 5}} f(x) = f(5) = 5$. As $x$ gets closer and closer to $5$ from the right and left, we see that $y$ approaches $5$, which is equivalent to $f(5)$. 

```python
# Define the function f(x) = 2x - 5
def f(x):
    return 2 * x - 5

# Generate x values
x = np.linspace(-10, 10, 400)  # 400 points between -10 and 10
# Generate y values using the function
y = f(x)
# Plot the function
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = 2x - 5', color='blue')
plt.scatter(5, 5, s=49, c='black')
plt.arrow(3, 2, 1, 2, head_width=0.3, head_length=1, 
          fc='red', ec='red')
plt.arrow(7, 10, -1, -2, head_width=0.3, head_length=1, 
          fc='red', ec='red')
# Add labels and title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(r'Depiction of $\lim_{{x \to 5}} \: f(x)$')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')  # x-axis
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')  # y-axis
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

However, limits are not always so straightforward. There are instances where $\lim_{{x \to a}} f(x) \neq f(a)$. See the below example where we have the function $f(x) = \frac{x^2 - 1}{x - 1}$ and the $\lim_{{x \to 1}} f(x)$. When solving for $f(1)$, we get the indeterminate form $\frac{0}{0}$ (good explanation of indeterminate forms [here](https://www.khanacademy.org/math/algebra/x2f8bb11595b61c86:foundation-algebra/x2f8bb11595b61c86:division-zero/v/undefined-and-indeterminate)). $1$ is the only value that will give us $\frac{0}{0}$, but any other value approaching $1$ from either side will give a real number. One potential method to find a limit of a rational function is to simplify:

$ \frac{x^2 - 1}{x - 1} = \frac{(x-1)(x+1)}{x-1}$
<br><br>
$\frac{(x-1)(x+1)}{x-1} = x+1$

When plugging $1$ in for $x+1$, we get $2$.

Thus, $f(1)$ is indeterminate, but the $\lim_{{x \to 1}} f(x) = 2$

```python
# Define the function with a removable discontinuity at x=1
def f(x):
    return (x**2 - 1) / (x - 1)

# Generate x values (excluding 1 to avoid division by zero)
x_left = np.linspace(-2, 0.95, 400)  # Values slightly less than 1
x_right = np.linspace(1.05, 2, 400)   # Values slightly greater than 1
x = np.concatenate((x_left, x_right))

# Generate y values using the function
y = f(x)

# Plot the function
plt.figure(figsize=(8, 6))
plt.plot(x_left, f(x_left), color='blue')
plt.plot(x_right, f(x_right), color='blue') 

# Add a blank circle at the gap (x=1)
plt.scatter(1, 2, color='black', facecolors='none', s=100, 
            edgecolors='black', label='Removable Discontinuity')

plt.arrow(0.6, 1.8, 0.2, 0.2, head_width=0.1, head_length=0.1, 
          fc='red', ec='red')
plt.arrow(1.4, 2.6, -0.2, -0.2, head_width=0.1, head_length=0.1, 
          fc='red', ec='red')

plt.text(-2, 4, r'$f(x) = \frac{x^2 - 1}{x - 1}$', fontsize=12, 
         bbox=dict(facecolor='white', alpha=0.3))


# Add labels and title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(r'Depiction of $\lim_{{x \to 1}} \: f(x)$ (removable discontinuity at x = 1)')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')  # x-axis
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')  # y-axis
plt.grid(True, linestyle='--', alpha=0.7)
plt.ylim(-5, 5)  # Set y-axis limits for better visualization
plt.show()
```

To further illustrate this distinction between $\lim_{{x \to a}} f(x)$ and $f(a)$, read the below code that uses $f(x)$ from the last example. As $x$ gets closer and closer to $1$ from the left and right, f gets closer and closer to $2$, but it will never arrive at $2$ until $x$ actually equals $1$. When dealing with limits, we care about what $f$ is approaching as $x$ approaches an arbitrary number $a$, not $f(a)$. Sometimes these things are equal and sometimes they aren't as we've observed above.

```python
for x in [0.9,0.99,0.999]:
    print(f'x = {x}, f(x) = {np.round(f(x),3)}')
print(50*'-')
for x in [1.1,1.01,1.001]:
    print(f'x = {x}, f(x) = {np.round(f(x),3)}')
```

### A Limit Must Be the Same from Both Sides

In the past two examples, we see the limit as $x$ approaches $a$ is the same coming from the left and right sides of the graph ($5$ in first example and $2$ in the second). If a function's limit from the left side isn't the same as the limit from the right side, then that limit doesn't exist. See the example in the graph below, which plots the piecewise function $f(x) = \begin{cases} x + 1 & \text{if } x < 2 \\x - 1 & \text{if } x \geq 2 \end{cases}$

```python
# Define the function with a jump discontinuity at x = 2
def jump_function(x):
    if x < 2:
        return x + 1
    else:
        return x - 1

# Generate x values
x = np.linspace(0, 4, 400)

# Generate y values using the jump_function
y = np.vectorize(jump_function)(x)

# Plot the function
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Jump Discontinuity at x = 2', color='b', linewidth=2)

# Add circles
plt.scatter(2, 3, color='black', facecolors='none', s=100, edgecolors='black')
plt.scatter(2, 1, color='black', s=100, edgecolors='black')

# Add labels and title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Graph of f(x) with a Jump Discontinuity at x = 2')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')  # x-axis
plt.axvline(2, color='red', linewidth=1.5, linestyle='--', label='x = 2 (Discontinuity Point)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

In the above graph, one can see that as $x$ approaches $2$ from the left, $f(x)$ approaches 3. However, as $x$ approaches $2$ from the right, $f(x)$ approaches 1. Since these are different values, the $\lim_{{x \to 2}} f(x)$ does not exist.

So far we've only looked at two-sided limits, but the concept of one-sided limits exist as well. Adding a $-$ sign indicates approaching from the left, while a $+$ sign indicates approaching from the right, like so: $\lim_{{x \to a^-}} f(x)$ and $\lim_{{x \to a^+}} f(x)$, respectively. In the above function, these one-sided limits *do* exist. $\lim_{{x \to 2^-}} f(x) = 3$ and $\lim_{{x \to 2^+}} f(x) = 1$, as further illustrated below:

```python
for x in [1.9,1.99,1.999]:
    print(f'x = {x}, f(x) = {np.round(jump_function(x),3)}')
print(50*'-')
for x in [2.1,2.01,2.001]:
    print(f'x = {x}, f(x) = {np.round(jump_function(x),3)}')
```

### Understanding the Squeeze Theorem

The Squeeze Theorem is a fundamental concept in calculus that helps us understand the behavior of functions when they are "squeezed" between two other functions. 

The Squeeze Theorem states that if $h(x) \leq g(x) \leq f(x)$ over some interval and both $h(x)$ and $f(x)$ approach the same limit as $x$ approaches $c$, then $g(x)$ must also converge to the same limit at $c$. This theorem sounds like common sense, but it's been helpful in finding the limits of some tricky functions.

To illustrate the theorem, refer to the graphs below. Three functions are plotted:

$f(x) = x^2 + 1$

$g(x) = \frac{sin(x)}{x}$

$h(x) = -x^2 + 1$

One can calculate via direct substitution that the $\lim_{{x \to 0}} f(x) = 1$ and $\lim_{{x \to 0}} h(x) = 1$. Since we know that $g(x)$ is sandwiched between $f(x)$ and $h(x)$, by the squeeze theorem we know that the $\lim_{{x \to 0}} g(x)$ also equals $1$, which is something we couldn't have figured out with direct substituion since it'd lead to an indeterminate form!

```python
# Define the functions
def f(x):
    return x**2 + 1
def g(x):
    return np.sin(x)/x
def h(x):
    return -x**2 + 1

# Generate x values
x = np.linspace(-1, 1, 400)

# Generate y values for the functions
y_f = f(x)
y_g = g(x)
y_h = h(x)

# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Plot the functions in the first subplot
axs[0].plot(x, y_f, label='f(x) = x^2 + 1', color='blue')
axs[0].plot(x, y_g, label='g(x) = sin(x)/x', color='green')
axs[0].plot(x, y_h, label='h(x) = -x^2 + 1', color='red')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')
axs[0].set_title('Illustration of Squeeze Theorem')
axs[0].axhline(0, color='black', linewidth=0.5, linestyle='--')  # x-axis
axs[0].axvline(0, color='black', linewidth=0.5, linestyle='--')  # y-axis
axs[0].grid(True, linestyle='--', alpha=0.7)
axs[0].legend()
axs[0].set_ylim(0, 2)  # Set y-axis limits for better visualization

# Plot the zoomed-in version in the second subplot
axs[1].plot(x, y_f, label='f(x) = x^2 + 1', color='blue')
axs[1].plot(x, y_g, label='g(x) = sin(x)/x', color='green')
axs[1].plot(x, y_h, label='h(x) = -x^2 + 1', color='red')
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')
axs[1].set_title('Zoomed-In Version')
axs[1].axhline(0, color='black', linewidth=0.5, linestyle='--')  # x-axis
axs[1].axvline(0, color='black', linewidth=0.5, linestyle='--')  # y-axis
axs[1].grid(True, linestyle='--', alpha=0.7)
axs[1].legend()
axs[1].set_ylim(0.75, 1.25)  # Set y-axis limits for better visualization
axs[1].set_xlim(-0.3, 0.3)  # Set x-axis limits for zoomed-in view

plt.tight_layout()
plt.show()
```
