# Introduction to Limits in Calculus

```python
import numpy as np
import matplotlib.pyplot as plt
```

In calculus, the concept of **limits** plays a fundamental role. Limits allow us to understand the behavior of functions as they approach specific points or values. 

Consider a function $f(x)$. The limit of $f(x)$ as $x$ approaches a certain value, say $a$, represents the value that $f(x)$ approaches as $x$ gets arbitrarily close to $a$.

Mathematically, we denote the limit as:

$$\lim_{{x \to a}} f(x)$$

This signifies the value that $f(x)$ approaches as $x$ approaches $a$.

Limits are essential for analyzing continuity, derivatives, and integrals in calculus, forming the basis for many advanced concepts in mathematics.

Limits can be straightforward, where $\lim_{{x \to a}} f(x) = f(a)$. See the below example where we have the function $f(x) = 2x - 5$. The $\lim_{{x \to 5}} f(x) = f(5) = 5$. As $x$ gets closer and closer to 5 from the right and left, we see that $y$ approaches 5, which is equivalent to $f(5)$. 

```python
# Define the function f(x) = 2x - 5
def f(x):
    return 2 * x - 5

# Generate x values
x = np.linspace(-10, 10, 400)  # 400 points between -10 and 10

# Generate y values using the function
y = f(x)

# Plot the function
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = 2x - 5', color='blue')
plt.scatter(5, 5, s=49, c='black')
plt.arrow(3, 2, 1, 2, head_width=0.3, head_length=1, 
          fc='red', ec='red')
plt.arrow(7, 10, -1, -2, head_width=0.3, head_length=1, 
          fc='red', ec='red')

# Add labels and title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x) = 2x - 5')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')  # x-axis
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')  # y-axis
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

However, limits are not always so straightforward. There are instances where $\lim_{{x \to a}} f(x) \neq f(a)$. See the below example where we have the function $f(x)

```python
# Define the function with a removable discontinuity at x=1
def f(x):
    return (x**2 - 1) / (x - 1)

# Generate x values (excluding 1 to avoid division by zero)
x_left = np.linspace(-2, 0.95, 400)  # Values slightly less than 1
x_right = np.linspace(1.05, 2, 400)   # Values slightly greater than 1
x = np.concatenate((x_left, x_right))

# Generate y values using the function
y = f(x)

# Plot the function
plt.figure(figsize=(8, 6))
plt.plot(x_left, f(x_left), color='blue')
plt.plot(x_right, f(x_right), color='blue')

# Add a blank circle at the gap (x=1)
plt.scatter(1, 2, color='black', facecolors='none', s=100, 
            edgecolors='black', label='Removable Discontinuity')

plt.arrow(0.6, 1.8, 0.2, 0.2, head_width=0.1, head_length=0.1, 
          fc='red', ec='red')
plt.arrow(1.4, 2.6, -0.2, -0.2, head_width=0.1, head_length=0.1, 
          fc='red', ec='red')

# Add labels and title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x) = (x^2 - 1) / (x - 1) (Removable Discontinuity at x = 1)')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')  # x-axis
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')  # y-axis
plt.grid(True, linestyle='--', alpha=0.7)
plt.ylim(-5, 5)  # Set y-axis limits for better visualization
plt.show()
```
