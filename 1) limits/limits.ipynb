```python
import numpy as np
import sympy as smp
import matplotlib.pyplot as plt
```

```python
def plot_fcn(fcn, fcn_title):
    '''Template for graphing mathematical functions'''
    # Generate x values from 1 to 10 (or any other appropriate range)
    x = np.linspace(-10, 10, 400)
    # Generate y values using the function f(x)
    y = fcn(x)
    # Plot the function
    plt.figure(figsize=(8, 6))
    plt.plot(x, y, color='blue', linewidth=2)
    
    # Add labels and title
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title(f'Illustration of {fcn_title}')
    plt.axhline(0, color='black', linewidth=0.5, linestyle='--')  # x-axis
    plt.axvline(0, color='black', linewidth=0.5, linestyle='--')  # y-axis
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.xlim(-10, 10)
    plt.ylim(-15, 15)  # Set y-axis limits for better visualization
    plt.show()
```

# Introduction to Limits in Calculus

The concept of **limits** plays a fundamental role in Calculus. Limits allow us to understand the behavior of functions as they approach specific points or values. 

Consider a function $f(x)$. The limit of $f(x)$ as $x$ approaches a certain value, say $a$, represents the value that $f(x)$ approaches as $x$ gets arbitrarily close to $a$.

Mathematically, we denote the limit as:

$$\lim_{{x \to a}} f(x)$$

This signifies the value that $f(x)$ approaches as $x$ approaches $a$.

Limits are essential for analyzing continuity, derivatives, and integrals in calculus, forming the basis for many advanced concepts in mathematics.

### Different kinds of limits

Limits can be straightforward, where $\lim_{{x \to a}} f(x) = f(a)$. See the below example where we have the function $f(x) = 2x - 5$. The $\lim_{{x \to 5}} f(x) = f(5) = 5$. As $x$ gets closer and closer to $5$ from the right and left, we see that $y$ approaches $5$, which is equivalent to $f(5)$. 

```python
# Define the function f(x) = 2x - 5
def f(x):
    return 2 * x - 5

# Generate x values
x = np.linspace(-10, 10, 400)  # 400 points between -10 and 10
# Generate y values using the function
y = f(x)
# Plot the function
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = 2x - 5', color='blue')
plt.scatter(5, 5, s=49, c='black')
plt.arrow(3, 2, 1, 2, head_width=0.3, head_length=1, 
          fc='red', ec='red')
plt.arrow(7, 10, -1, -2, head_width=0.3, head_length=1, 
          fc='red', ec='red')
# Add labels and title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(r'Depiction of $\lim_{{x \to 5}} \: f(x)$')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')  # x-axis
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')  # y-axis
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

```python
# showing 2x-5 limit solution using sympy library
x = smp.symbols('x')
smp.limit(2*x-5, x, 5)
```

However, limits are not always so straightforward. There are instances where $\lim_{{x \to a}} f(x) \neq f(a)$. See the below example where we have the function $f(x) = \frac{x^2 - 1}{x - 1}$ and the $\lim_{{x \to 1}} f(x)$. When solving for $f(1)$, we get the indeterminate form $\frac{0}{0}$ (good explanation of indeterminate forms [here](https://www.khanacademy.org/math/algebra/x2f8bb11595b61c86:foundation-algebra/x2f8bb11595b61c86:division-zero/v/undefined-and-indeterminate)). $1$ is the only value that will give us $\frac{0}{0}$, but any other value approaching $1$ from either side will give a real number. One potential method to find a limit of a rational function is to simplify:

$ \frac{x^2 - 1}{x - 1} = \frac{(x-1)(x+1)}{x-1}$
<br><br>
$\frac{(x-1)(x+1)}{x-1} = x+1$

When plugging $1$ in for $x+1$, we get $2$.

Thus, $f(1)$ is indeterminate, but the $\lim_{{x \to 1}} f(x) = 2$

Note: one could say that $\frac{x^2 - 1}{x - 1} = x + 1$, where $x \neq 1$

```python
# Define the function with a removable discontinuity at x=1
def f(x):
    return (x**2 - 1) / (x - 1)

# Generate x values (excluding 1 to avoid division by zero)
x_left = np.linspace(-2, 0.95, 400)  # Values slightly less than 1
x_right = np.linspace(1.05, 2, 400)   # Values slightly greater than 1
x = np.concatenate((x_left, x_right))

# Generate y values using the function
y = f(x)

# Plot the function
plt.figure(figsize=(8, 6))
plt.plot(x_left, f(x_left), color='blue')
plt.plot(x_right, f(x_right), color='blue') 

# Add a blank circle at the gap (x=1)
plt.scatter(1, 2, color='black', facecolors='none', s=100, 
            edgecolors='black', label='Removable Discontinuity')

plt.arrow(0.6, 1.8, 0.2, 0.2, head_width=0.1, head_length=0.1, 
          fc='red', ec='red')
plt.arrow(1.4, 2.6, -0.2, -0.2, head_width=0.1, head_length=0.1, 
          fc='red', ec='red')

plt.text(-2, 4, r'$f(x) = \frac{x^2 - 1}{x - 1}$', fontsize=12, 
         bbox=dict(facecolor='white', alpha=0.3))


# Add labels and title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(r'Depiction of $\lim_{{x \to 1}} \: f(x)$ (removable discontinuity at x = 1)')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')  # x-axis
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')  # y-axis
plt.grid(True, linestyle='--', alpha=0.7)
plt.ylim(-5, 5)  # Set y-axis limits for better visualization
plt.show()
```

```python
# verifying solution with sympy
x = smp.symbols('x')
smp.limit((x**2 - 1)/(x-1), x, 1)
```

To further illustrate this distinction between $\lim_{{x \to a}} f(x)$ and $f(a)$, read the below code that uses $f(x)$ from the last example. As $x$ gets closer and closer to $1$ from the left and right, f gets closer and closer to $2$, but it will never arrive at $2$. When dealing with limits, we care about what $f$ is approaching as $x$ approaches an arbitrary number $a$, not $f(a)$. Sometimes these things are equal and sometimes they aren't as we've observed above.

```python
for x in [0.9,0.99,0.999]:
    print(f'x = {x}, f(x) = {np.round(f(x),3)}')
print(50*'-')
for x in [1.1,1.01,1.001]:
    print(f'x = {x}, f(x) = {np.round(f(x),3)}')
```

When we actually try to input $1$ for $f(x)$, we get an error:

```python
try:
    f(1)  # This will raise a ZeroDivisionError
except ZeroDivisionError as e:
    # Handle the ZeroDivisionError exception
    print("Error: Division by zero!")
```

### A Limit Must Be the Same from Both Sides

In the past two examples, we see the limit as $x$ approaches $a$ is the same coming from the left and right sides of the graph ($5$ in first example and $2$ in the second). If a function's limit from the left side isn't the same as the limit from the right side, then that limit doesn't exist. See the example in the graph below, which plots the piecewise function $f(x) = \begin{cases} x + 1 & \text{if } x < 2 \\x - 1 & \text{if } x \geq 2 \end{cases}$

```python
# Define the function with a jump discontinuity at x = 2
def jump_function(x):
    if x < 2:
        return x + 1
    else:
        return x - 1

# Generate x values
x = np.linspace(0, 4, 400)

# Generate y values using the jump_function
y = np.vectorize(jump_function)(x)

# Plot the function
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Jump Discontinuity at x = 2', color='b', linewidth=2)

# Add circles
plt.scatter(2, 3, color='black', facecolors='none', s=100, edgecolors='black')
plt.scatter(2, 1, color='black', s=100, edgecolors='black')

# Add labels and title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Graph of f(x) with a Jump Discontinuity at x = 2')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')  # x-axis
plt.axvline(2, color='red', linewidth=1.5, linestyle='--', label='x = 2 (Discontinuity Point)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

In the above graph, one can see that as $x$ approaches $2$ from the left, $f(x)$ approaches 3. However, as $x$ approaches $2$ from the right, $f(x)$ approaches 1. Since these are different values, the $\lim_{{x \to 2}} f(x)$ does not exist.

So far we've only looked at two-sided limits, but the concept of one-sided limits exist as well. Adding a $-$ sign indicates approaching from the left, while a $+$ sign indicates approaching from the right, like so: $\lim_{{x \to a^-}} f(x)$ and $\lim_{{x \to a^+}} f(x)$, respectively. In the above function, these one-sided limits *do* exist. $\lim_{{x \to 2^-}} f(x) = 3$ and $\lim_{{x \to 2^+}} f(x) = 1$, as further illustrated below:

```python
for x in [1.9,1.99,1.999]:
    print(f'x = {x}, f(x) = {np.round(jump_function(x),3)}')
print(50*'-')
for x in [2.1,2.01,2.001]:
    print(f'x = {x}, f(x) = {np.round(jump_function(x),3)}')
```

### Understanding the Squeeze Theorem

The Squeeze Theorem is a fundamental concept in calculus that helps us understand the behavior of functions when they are "squeezed" between two other functions. 

The Squeeze Theorem states that if $h(x) \leq g(x) \leq f(x)$ over some interval and both $h(x)$ and $f(x)$ approach the same limit as $x$ approaches $c$, then $g(x)$ must also converge to the same limit at $c$. This theorem sounds like common sense, but it's proven helpful in finding the limits of some tricky functions.

To illustrate the theorem, refer to the graphs below. Three functions are plotted:

$f(x) = x^2 + 1$

$g(x) = \frac{sin(x)}{x}$

$h(x) = -x^2 + 1$

One can calculate via direct substitution that the $\lim_{{x \to 0}} f(x) = 1$ and $\lim_{{x \to 0}} h(x) = 1$. Since we know that $g(x)$ is sandwiched between $f(x)$ and $h(x)$, by the squeeze theorem we know that the $\lim_{{x \to 0}} g(x)$ also equals $1$, which is something we couldn't have figured out with direct substituion since it'd lead to an indeterminate form!

```python
# Define the functions
def f(x):
    return x**2 + 1
def g(x):
    return np.sin(x)/x
def h(x):
    return -x**2 + 1

# Generate x values
x = np.linspace(-1, 1, 400)

# Generate y values for the functions
y_f = f(x)
y_g = g(x)
y_h = h(x)

# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Plot the functions in the first subplot
axs[0].plot(x, y_f, label=r'$f(x) = x^2 + 1$', color='blue')
axs[0].plot(x, y_g, label=r'$g(x) = \frac{sin(x)}{x}$', color='green')
axs[0].plot(x, y_h, label=r'$h(x) = -x^2 + 1$', color='red')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')
axs[0].set_title('Illustration of Squeeze Theorem')
axs[0].axhline(0, color='black', linewidth=0.5, linestyle='--')  # x-axis
axs[0].axvline(0, color='black', linewidth=0.5, linestyle='--')  # y-axis
axs[0].grid(True, linestyle='--', alpha=0.7)
axs[0].legend()
axs[0].set_ylim(0, 2)  # Set y-axis limits for better visualization

# Plot the zoomed-in version in the second subplot
axs[1].plot(x, y_f, label=r'$f(x) = x^2 + 1$', color='blue')
axs[1].plot(x, y_g, label=r'$g(x) = \frac{sin(x)}{x}$', color='green')
axs[1].plot(x, y_h, label=r'$h(x) = -x^2 + 1$', color='red')
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')
axs[1].set_title('Zoomed-In Version')
axs[1].axhline(0, color='black', linewidth=0.5, linestyle='--')  # x-axis
axs[1].axvline(0, color='black', linewidth=0.5, linestyle='--')  # y-axis
axs[1].grid(True, linestyle='--', alpha=0.7)
axs[1].legend()
axs[1].set_ylim(0.75, 1.25)  # Set y-axis limits for better visualization
axs[1].set_xlim(-0.3, 0.3)  # Set x-axis limits for zoomed-in view

plt.tight_layout()
plt.show()
```

```python
# verifying three function solutions are equal, with sympy
x = smp.symbols('x')
print(smp.limit(x**2 +1, x, 0))
print(smp.limit(-x**2 +1, x, 0))
print(smp.limit(smp.sin(x)/x, x, 0))
```

### Infinite Limits

Limits can also approach $\infty$ and $-\infty$. Starting with a simple example of $f(x) = \frac{1}{x}$, we can find that the limits as x approaches $\infty$ and $-\infty$ are both 0 by plugging in larger and larger positive and negative numbers:

```python
# as x approaches infinity, f(x) approaches 0 from the positive side
for x in [10, 100, 1_000, 10_000]:
    print(1/x)
print('--'*50)

# as x approaches negative infinity, f(x) approaches 0 from the negative side
for x in [-10, -100, -1_000, -10_000]:
    print(1/x)
```

Refer to the below graph for help visualizing the $\lim_{{x \to \infty}} \frac{1}{x}$ and $\lim_{{x \to -\infty}} \frac{1}{x}$:

```python
# horizontal asymptotes at 0 form as x approaches + and - inf
def f_one_over_x(x):
    return 1 / x
plot_fcn(f_one_over_x, r'$f(x) = \frac{1}{x}$')
```

```python
# verifying solutions for lim of 1/x as x approaches + and - inf
x = smp.symbols('x')
print(smp.limit((1)/(x), x, smp.oo))
print(smp.limit((1)/(x), x, -smp.oo))
```

With longer rational functions, it can often be cumbersome to find $\lim_{{x \to \infty}}$ and $\lim_{{x \to -\infty}}$ by plugging in larger positive and negative numbers for x, such as with $f(x) = \frac{5x^7 + 100x^4 - 23x}{x^7 + 30.5x^4 - 400x^2}$

An easier way to calculate limits in these rational cases is to check the highest degree of x in the numerator and denominator. For instance, when looking at the numerator, a very large number to the 7th power is going to be much, much larger than that number to the 4th or 1st power, which will effectively render the latter two numbers irrelevant. The same logic goes for the denominator. So, when taking $\lim_{{x \to \infty}} \frac{5x^7 + 100x^4 - 23x}{x^7 + 30.5x^4 - 400x^2}$, one should look for $\lim_{{x \to \infty}} \frac{5x^7}{x^7}$. The $x^7$ in the numerator and denominator would cancel out to leave $\frac{5}{1}$. The $\lim_{{x \to \infty}} 5$ is just $5$!

If you tried the same exercise with $\lim_{{x \to -\infty}}\frac{5x^7 + 100x^4 - 23x}{x^7 + 30.5x^4 - 400x^2}$, we'd get the same result. The below graph shows how $f(x)$ approaches $5$ as $x$ approaches $\infty$ and $-\infty$:

```python
# horizontal asymptotes at 5 form as x approaches + and - inf
def f_rational(x):
    return (5*x**7 + 100*x**4 - 23*x)/(x**7 + 30.5*x**4 - 400*x**2)
plot_fcn(f_rational, r'$f(x) = \frac{5x^7 + 100x^4 - 23x}{x^7 + 30.5x^4 - 400x^2}$')
```

```python
# verifying solutions for lim of ratl function as x approaches + and - inf
x = smp.symbols('x')
print(smp.limit((5*x**7 + 100*x**4 - 23*x)/(x**7 + 30.5*x**4 - 400*x**2), x, smp.oo))
print(smp.limit((5*x**7 + 100*x**4 - 23*x)/(x**7 + 30.5*x**4 - 400*x**2), x, -smp.oo))
```

### Intermediate Value Theorem

The Intermediate Value Theorem (IVT) states that, assuming we have a continuous function within a closed interval $[a, b]$, for any $L$ between $f(a)$ and $f(b)$ values, there exists a $c$ between $[a, b]$ where $f(c) = L$. This is an intuitive theorem, basically saying that when we draw a line from $a$ to $b$ without picking up our pencil, we will cross all points ($L$) between $f(a)$ to $f(b)$.

Refer to the below graph for a visual demonstration. We have the function $f(x) = x^3 - 4x^2 + x + 6$ over the closed interval $[1,3]$. The code selects a $y$ (aka $L$) value at random falling between $f(1)$ and $f(3)$, then uses np.roots method to calculate the corresponding $x$ value (aka $c$). No matter how many times one runs the cell, $c$ falls within $[a,b]$!

```python
# Define the function (one can change this to any continuous function)
def f(x):
    return x**3 - 4*x**2 + x + 6

# Define the closed interval [a, b]
a = 1
b = 3

# Calculate function values at a and b
fa = f(a)
fb = f(b)

# Choose a random y-value between f(a) and f(b)
y = np.round(np.random.uniform(fa, fb),2)

# Define the coefficients of the polynomial equation f(x) = y --> f(x) - y = 0
coefficients = [1, -4, 1, 6 - y]
# Solve for x when f(x) = y using NumPy's roots function
roots = np.roots(coefficients)
# Select a root within the interval [a, b]
c = [root for root in roots if a <= root <= b][0]

# Plot function
x = np.linspace(0, 4, 400)
y_values = f(x)
plt.figure(figsize=(8, 6))
plt.plot(x, y_values, color='blue', linewidth=2)

# Highlight the points (a, f(a)) and (b, f(b)) in red
plt.scatter([a, b], [fa, fb], color='red', zorder=5)
plt.text(a, fa, f'({a}, {fa})', fontsize=10, verticalalignment='bottom', horizontalalignment='right')
plt.text(b, fb, f'({b}, {fb})', fontsize=10, verticalalignment='bottom', horizontalalignment='right')

# Highlight the point (c, f(c)) in green
plt.scatter([c], [y], color='green', zorder=5)
plt.text(c, y, f'({c:.2f}, {y})', fontsize=10, verticalalignment='bottom', horizontalalignment='right')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Intermediate Value Theorem Illustration')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')  # x-axis
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')  # y-axis
plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```
